# 灵敏度分析及图片导出
- 分析的算法参数：$C_E$, $\alpha$, $\beta$
    - $C_E$: $\{10^0, 10^{-2}, 10^{-4}, \dots, 10^{-64}\}$
    - $\alpha$: $\{0.01, 0.02, 0.03, \dots, 0.20\}$
    - $\beta$: $\{0.1, 0.2, 0.3, \dots, 1.0\}$
    - 控制变量用的其他参数：$C_E=10^{-32}, \alpha=0.02, \beta=0.8$

- 图片导出的设定
    - 图片数量：$3 \times 3$, 其中 3 个参数，3 个指标
    - 图片位置：同一参数图片 $3 \times 1$
    - 图片大小：$3 \times 1$ 图片尺寸为 `width=16cm`, `height=5cm`
    - 图片字体大小：可选大小为 `7pt, 8pt, 9pt, 10pt`

```python
%pylab inline
import pandas as pd
import numpy as np
```

```python
dfcapture = pd.read_csv('Exps/sensitivity_analysis_capture.csv', header=None)
dfalpha = pd.read_csv('Exps/sensitivity_analysis_alpha.csv', header=None)
dfbeta = pd.read_csv('Exps/sensitivity_analysis_beta.csv', header=None)
```

```python
dfcapture = dfcapture.T
dfalpha   = dfalpha.T
dfbeta    = dfbeta.T
df = [dfcapture, dfalpha, dfbeta]
```

```python
sa_col = ['capture', 'alpha', 'beta']
Tsafe_cols = list(range(0, 99, 3))
TR_cols = list(range(1, 99, 3))
Ei_cols = list(range(2, 99, 3))

capture_cols = [r'$10^{'+str(x)+'}$' if (i) % 8 == 0 else ''
                for i, x in enumerate(np.log10(np.array(list((dfcapture.iloc[3][::3].values)))).astype(int))]
alpha_cols   = [str(x) if (i)%5 == 0 else ''
                for i, x in enumerate(np.array(list((dfalpha.iloc[1][::3].values))))]
beta_cols = [str(x) if (i)%5 == 0 else ''
             for i, x in enumerate(np.array(list((dfbeta.iloc[2][::3].values))))]

performance_list = [Tsafe_cols, TR_cols, Ei_cols]
parameter_list = [capture_cols, alpha_cols, beta_cols]
```

# 图片参数
- 数据名
- 图像参数

```python
col_names_zh = [u'安全周期 ('+r'$10^3$'+')', u'传输时延 ('+r'$log$ '+'ns)', u'网络能耗 ('+r'$10^5$ '+'nJ)']
alg_names_zh = [u'幻影路由', u'诱导树路由', u'动态虚假源选择', u'虚假源调度']
subtitles_zh = [u'安全周期与参数 ', u'传输时延与参数 ', u'网络能耗与参数 ']
col_names_en = ['Safety ('+r'$10^3$'+')', 'Delay ('+r'$log$ '+'ns)', 'Consumption ('+r'$10^5$ '+'nJ)']
subtitles_en = ['The safety under different ', 'The delay under different ', 'The energy consumption under different ']
subtitles_ids = ['(a) ', '(b) ', '(c) ']
para_names = [r'$C_E$', r'$\alpha$', r'$\beta$']
figformat = ['png', 'eps', 'pdf']
sa_names = ['safety', 'delay', 'energy_consumption']
import matplotlib.font_manager as fm
zhfont = fm.FontProperties(fname='C:/Windows/Fonts/msyh.ttc')#(fname='/home/sturui/.local/share/fonts/simsun.ttf')
matplotlib.rcParams.update({'font.size':10, 'font.family':'serif'})
```

# 存储图片

```python
j2y = [([0, 1e3], ['0', '1']),
      ([1e7, 1e8, 1e9, 1e10, 1e11], ['7', '8', '9', '10', '11']),
      ([5e5, 7e5, 9e5, 11e5], ['5', '7', '9', '11']),
      ([5e5, 10e5, 15e5, 20e5, 25e5, 30e5], ['5', '10', '15', '20', '25','30']),
      ([5e5, 7e5, 9e5, 11e5, 13e5], ['5', '7', '9', '11', '13'])]
for i, data in enumerate(df):
    plt.figure(figsize=(8.1, 2))
    for j, col in enumerate(performance_list):
        col = list(range(j, max(data.columns)+1, 3))
        dat = [list(data[d][4:].values) for d in col]
        plt.subplot(1,3,j+1)
        plt.boxplot(dat, labels=parameter_list[i])
        plt.plot(list(range(1, len(parameter_list[i])+1)), [np.mean(d) for d in dat], lw=1, color='red', marker='p')
        if j == 1:# or j == 2:
            plt.yscale('log')

        plt.yticks(j2y[j][0], j2y[j][1])
        if i > 0 and j == 2:
            plt.yticks(j2y[j+i][0], j2y[j+i][1])
        plt.yticks(fontsize=8)
        plt.xticks(fontsize=8)
        
        ax = plt.gca()
        ax.set_xlabel(para_names[i], fontsize=8)
        # china figure
#         ax.set_ylabel(col_names_zh[j], fontproperties=zhfont, fontsize=8)
#         ax.set_title(subtitles_ids[j]+subtitles_zh[j]+para_names[i]+u' 的关系', 
#                      fontproperties=zhfont, position=(0.5, -0.33), fontsize=9)
        # english figure
        ax.set_ylabel(col_names_en[j], fontsize=8)
        ax.set_title(subtitles_ids[j]+subtitles_en[j]+para_names[i], 
                     position=(0.5, -0.32), fontsize=9)
        # xlabel, ylabel 位置设定
        ax.xaxis.set_label_coords(0.5, -0.12)
        ax.yaxis.set_label_coords(-0.1, 0.5)
    ff = figformat[2]
    fig = plt.gcf()
    # english paper
    fig.subplots_adjust(top=0.95, bottom=0.22, left=0.045, right=0.93, wspace=0.25) # left, right, bottom, top, wspace, hspace
    # china paper ***
    fig.savefig('Figures/Sensitivity_Analysis/en/'+sa_col[i]+'_'+sa_names[i]+'.'+ff, dpi=600, format=ff)
    plt.show()
```
